```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Numerical columns in the DataFrame:")
for col in numerical_columns:
    print(col)
```

```python
correlation_matrix = df[numerical_columns].corr()
print("Head of the correlation matrix:")
print(correlation_matrix.head())
```

```python
strong_correlations = []
threshold = 0.7

for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        feature1 = correlation_matrix.columns[i]
        feature2 = correlation_matrix.columns[j]
        correlation = correlation_matrix.iloc[i, j]
        abs_correlation = abs(correlation)

        if abs_correlation > threshold:
            strong_correlations.append({
                'Feature 1': feature1,
                'Feature 2': feature2,
                'Correlation': correlation,
                'Absolute Correlation': abs_correlation
            })

strong_correlations_df = pd.DataFrame(strong_correlations)
strong_correlations_df = strong_correlations_df.sort_values(by='Absolute Correlation', ascending=False)

print("Strongly correlated pairs (Absolute Correlation > 0.7):")
print(strong_correlations_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the features with the strongest absolute correlation
strongest_correlation_pair = strong_correlations_df.iloc[0]
feature1_name = strongest_correlation_pair['Feature 1']
feature2_name = strongest_correlation_pair['Feature 2']

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=feature1_name, y=feature2_name, data=df, alpha=0.5)
plt.title(f'Scatter Plot of {feature1_name} vs. {feature2_name}')
plt.xlabel(feature1_name)
plt.ylabel(feature2_name)
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Helper function to determine the deck combo string
def get_deck_combo(card_list, xbow_id, guards_id, heal_spirit_id, graveyard_id):
    # Ensure xbow_id is present, as this function should only be called for X-bow decks
    if xbow_id not in card_list:
        return "Non-X-bow_Deck" # Should ideally not happen if filtering is correct

    has_guards = guards_id in card_list
    has_heal_spirit = heal_spirit_id in card_list
    has_graveyard = graveyard_id in card_list

    # Count how many of the specific companion cards are present
    companion_count = sum([has_guards, has_heal_spirit, has_graveyard])

    if companion_count == 0:
        return "X-bow-only"
    elif companion_count == 1:
        if has_guards:
            return "X-bow + Guards"
        elif has_heal_spirit:
            return "X-bow + Heal Spirit"
        elif has_graveyard:
            return "X-bow + Graveyard"
    else: # companion_count > 1, means multiple companions are present
        return "Other X-bow Combo" # This will capture 3-card combos and 2-card combos not explicitly named

# Function to process a DataFrame segment (df_before or df_after)
def process_xbow_decks(df_segment, period_name, xbow_id, guards_id, heal_spirit_id, graveyard_id):
    records = []

    for _, row in df_segment.iterrows():
        # Process winner's deck if it contains X-bow
        if xbow_id in row['winner.cards.list']:
            deck_info = {
                'period': period_name,
                'is_winner': 1, # 1 for winner
                'deck_combo': get_deck_combo(row['winner.cards.list'], xbow_id, guards_id, heal_spirit_id, graveyard_id),
                'totalcard.level': row['winner.totalcard.level'],
                'elixir.average': row['winner.elixir.average'],
                'startingTrophies': row['winner.startingTrophies'],
                'trophyChange': row['winner.trophyChange'],
                'crowns': row['winner.crowns'],
                'troop.count': row['winner.troop.count'],
                'structure.count': row['winner.structure.count'],
                'spell.count': row['winner.spell.count'],
                'common.count': row['winner.common.count'],
                'rare.count': row['winner.rare.count'],
                'epic.count': row['winner.epic.count'],
                'legendary.count': row['winner.legendary.count']
            }
            records.append(deck_info)

        # Process loser's deck if it contains X-bow
        if xbow_id in row['loser.cards.list']:
            deck_info = {
                'period': period_name,
                'is_winner': 0, # 0 for loser
                'deck_combo': get_deck_combo(row['loser.cards.list'], xbow_id, guards_id, heal_spirit_id, graveyard_id),
                'totalcard.level': row['loser.totalcard.level'],
                'elixir.average': row['loser.elixir.average'],
                'startingTrophies': row['loser.startingTrophies'],
                'trophyChange': row['loser.trophyChange'],
                'crowns': row['loser.crowns'],
                'troop.count': row['loser.troop.count'],
                'structure.count': row['loser.structure.count'],
                'spell.count': row['loser.spell.count'],
                'common.count': row['loser.common.count'],
                'rare.count': row['loser.rare.count'],
                'epic.count': row['loser.epic.count'],
                'legendary.count': row['loser.legendary.count']
            }
            records.append(deck_info)

    return pd.DataFrame(records)

# Process df_before
df_xbow_archetypes_before = process_xbow_decks(df_before, 'before', xbow_card_id, guards_card_id, heal_spirit_card_id, graveyard_card_id)

# Process df_after
df_xbow_archetypes_after = process_xbow_decks(df_after, 'after', xbow_card_id, guards_card_id, heal_spirit_card_id, graveyard_card_id)

# Concatenate the results
df_xbow_archetypes = pd.concat([df_xbow_archetypes_before, df_xbow_archetypes_after], ignore_index=True)

print("Head of df_xbow_archetypes:")
display(df_xbow_archetypes.head())

display(df_xbow_archetypes.tail())
```

```python
df_xbow_archetypes_encoded = pd.get_dummies(df_xbow_archetypes, columns=['deck_combo', 'period'], drop_first=True)
print("Head of df_xbow_archetypes_encoded after one-hot encoding:")
display(df_xbow_archetypes_encoded.head())
```

```python
import statsmodels.api as sm

# Define the dependent variable (is_winner)
y = df_xbow_archetypes_encoded['is_winner']

# Define the independent variables (deck_combo, period dummies, and elixir.average)
# Select one-hot encoded columns for deck_combo and period, and include elixir.average
feature_columns = [col for col in df_xbow_archetypes_encoded.columns if col.startswith('deck_combo_') or col.startswith('period_')]
# Add 'elixir.average' to the list of features
feature_columns.append('elixir.average')
X = df_xbow_archetypes_encoded[feature_columns].copy()

# Ensure all features are numerical. Boolean columns are converted to int.
for col in feature_columns:
    if X[col].dtype == 'bool':
        X[col] = X[col].astype(int)

# Add a constant (intercept) to the independent variables
X = sm.add_constant(X)

# Create and fit the Logit (Logistic Regression) model
model = sm.Logit(y, X)
results = model.fit()

# Print the model summary
print(results.summary())
```
